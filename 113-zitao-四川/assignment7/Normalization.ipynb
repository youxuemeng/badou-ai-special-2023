```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Normalization1(x):
    y = []
    for i in x:
       y.append((float(i) - min(x))/(max(x)-min(x)))
    return y

def Normalization2(x):
    y = []
    for i in x:
       y.append((float(i) - np.mean(x))/(max(x)-min(x)))
    return y

def z_score(x):
    x_mean = np.mean(x)
    s2 = sum([(i-x_mean) ** 2 for i in x]) / len(x)
    return [(i-x_mean)/s2 for i in x]
```

```python
l=[-10, 5, 5, 6, 6, 6, 7, 7, 7, 7, 8, 8, 8, 8, 8, 9, 9, 9, 9, 9, 9, 10, 10, 10, 10, 10, 10, 10, 11, 11, 11, 11, 11, 11, 12, 12, 12, 12, 12, 13, 13, 13, 13, 14, 14, 14, 15, 15, 30]

num = []
for i in l:
    c = l.count(i)
    num.append(c)

n = Normalization2(l)
z = z_score(l)
print(n)
print(z)

plt.plot(l, num)
plt.plot(z, num)
plt.show()
```
